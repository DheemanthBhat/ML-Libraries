#### Load Seaborn along with other libraries

```python
# Data manipulation libraries.
import numpy as np
import pandas as pd

# Data visualization libraries.
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns


print("Numpy version: ", np.__version__)
print("Pandas version: ", pd.__version__)
print("Matplotlib version: ", matplotlib.__version__)
print("Seaborn version: ", sns.__version__)
```

### Import Dataset

`seaborn` library has built in datasets. To get the list of datasets:

```python
print(sns.get_dataset_names())
```

Let use _car_crashes_ dataset.

```python
cc_df = sns.load_dataset("car_crashes")
cc_df.head(3)
```

## Styling

### Settings styles to axis and grids.

#### Without styling

```python
sns.jointplot(x="speeding", y="alcohol", data=cc_df, kind="reg")  # "reg" => "Regression"
plt.show()
```

#### With styling

Configure `seaborn` with:

1. White background with grid.
2. Set font scale to 1.5.

```python
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
```

#### _Joint Plot_ to show above changes

```python
sns.jointplot(x="speeding", y="alcohol", data=cc_df, kind="reg", height=8)
sns.despine(left=True, bottom=True)

plt.show()
```

### Palettes

Check [documentation][1] to get list of available color maps.

[1]: https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html#sequential

#### _Pair Plot_ to demonstrate _Palettes_

```python
tips_df = sns.load_dataset("tips")

sns.pairplot(tips_df, hue="sex", height=3, aspect=1.5, palette="inferno")
plt.show()
```
