#### Load Seaborn along with other libraries

```python
# Data manipulation libraries.
import numpy as np
import pandas as pd

# Data visualization libraries.
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns


print("Numpy version: ", np.__version__)
print("Pandas version: ", pd.__version__)
print("Matplotlib version: ", matplotlib.__version__)
print("Seaborn version: ", sns.__version__)
```

### Import Dataset

`seaborn` library has built in datasets. To get the list of datasets:

```python
print(sns.get_dataset_names())
```

Let use _car_crashes_ dataset.

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

### Configure Seaborn

```python
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
```

### Legend Locations

1. Position of Legend in a plot can be altered using `loc` (Location code) parameter of `plt.legend()`. 
2. Default value of `loc` is zero or "_best_".

| Location String | Location Code |
|--|--|
| best | 0 |
| upper right | 1 |
| upper left | 2 |
| lower left | 3 |
| lower right | 4 |
| right | 5 |
| center left | 6 |
| center right | 7 |
| lower center | 8 |
| upper center | 9 |
| center | 10 |

Check [documentation][1] for more details.

[1]: https://matplotlib.org/stable/api/legend_api.html#matplotlib.legend.Legend

## Categorical Plots

Types of [Categorical Plots](https://seaborn.pydata.org/api.html#categorical-plots):

1. Bar Plot
2. Count Plot
3. Box Plot
3. Violin Plot
5. Strip Plot
6. Swarm Plot

### 1. Bar Plot

```python
sns.barplot(x="sex", y="total_bill", data=tips_df)
plt.plot()
```

#### Setting an `estimator`

There are different types of estimators:
1. `np.median`: Median
2. `np.std`: Standard Deviation
3. `np.var`: Variance
4. `np.cov`: Co-Variance

```python
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=np.std)
plt.plot()
```

### 2. Count Plot

```python
sns.countplot(x="sex", data=tips_df)
plt.plot()
```

### 3. Box Plot

```python
plt.figure(figsize=[10, 6])
sns.boxplot(x="day", y="total_bill", data=tips_df, hue="sex")
plt.legend(loc=2)

plt.show()
```

### 4. Violin Plot

```python
plt.figure(figsize=[10, 6])
sns.violinplot(x="day", y="total_bill", data=tips_df, hue="sex")
plt.legend(loc=2)

plt.show()
```

### 5. Strip Plot

```python
plt.figure(figsize=[10, 6])
sns.stripplot(x="day", y="total_bill", data=tips_df, jitter=True, hue="sex", dodge=True)
plt.legend(loc=2)

plt.show()
```

1. `jitter`: Spreads out the data-points to avoid overlapping.
2. `hue`: Sets one color per category.
3. `dodge`: Each category gets separate lane.

### 6. Swarm Plot

```python
plt.figure(figsize=[10, 6])
sns.swarmplot(x="day", y="total_bill", data=tips_df, hue="sex", dodge=True)

plt.show()
```
