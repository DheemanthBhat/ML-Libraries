#### Load Seaborn along with other libraries

```python
# Data manipulation libraries.
import numpy as np
import pandas as pd

# Data visualization libraries.
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns


print("Numpy version: ", np.__version__)
print("Pandas version: ", pd.__version__)
print("Matplotlib version: ", matplotlib.__version__)
print("Seaborn version: ", sns.__version__)
```

### Import Dataset

`seaborn` library has built in datasets. To get the list of datasets:

```python
print(sns.get_dataset_names())
```

Let use _car_crashes_ dataset.

```python
cc_df = sns.load_dataset("car_crashes")
cc_df.head()
```

### Configure Seaborn

```python
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.5)
```

## Matrix Plots

Types of [Matrix Plots](https://seaborn.pydata.org/api.html#matrix-plots):

1. Heat Map
2. Cluster Map
3. Pair Grid
3. Facet Grid

### 1. Heat Map

#### Example 1: Car crash data

```python
# Get co-relaion matrix.
crash_mx = cc_df.corr()

plt.figure(figsize=[8, 6])
sns.heatmap(crash_mx, annot=True, cmap="Blues")

plt.plot()
```

#### Example 2: Flight data

**Step 1**: Load data

```python
flights = sns.load_dataset("flights")
flights.head()
```

**Step 2**: Create matrix

```python
flight_mx = flights.pivot_table(index="month", columns="year", values="passengers")
flight_mx
```

**Step 3**: Plot Heat Map

```python
plt.figure(figsize=[8, 6])
sns.heatmap(flight_mx, cmap="YlGnBu", linecolor="white", linewidth=1)

plt.show()
```

### 2. Cluster Map

```python
sns.clustermap(data=flight_mx, cmap="YlGnBu", standard_scale=1)
plt.show()
```

### 3. Pair Grid

```python
iris_df = sns.load_dataset("iris")

iris_g = sns.PairGrid(iris_df, hue="species", height=3)
iris_g.map_diag(plt.hist)
# iris_g.map_offdiag(plt.scatter)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)

plt.show()
```
