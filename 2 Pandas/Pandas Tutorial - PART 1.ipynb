## Introduction

[`pandas`](https://pandas.pydata.org/) is a software library written for the Python programming language for data manipulation and analysis. In particular, **it offers data structures and operations for manipulating numerical tables and time series**. The name is derived from the term "_panel data_", an econometrics term for data sets that include observations over multiple time periods for the same individuals.

Important data types used in `pandas` are:

1. Series
2. DataFrame

### Options

Some important options in `pandas` to alter the output format:

1. display.max_columns
2. display.max_rows
3. display.width

Syntax:

```python
pd.options.display.max_columns = 20
pd.options.display.max_rows = 10
pd.options.display.width = 130
```

#### Modify option `display.max_columns` to display required number of columns

````python
old_max_cols = pd.options.display.max_columns
print("Old value in option `display.max_columns`:", old_max_cols)

(row_count, col_count) = df.shape
pd.options.display.max_columns = col_count
print("New value in option `display.max_columns`:", col_count)
````

Disply first row to see all the columns.

````python
df.head(1)
````

### Schema of the _Stackoverflow_ data

_survey_results_schema.csv_ file contains details on each minified-column names in the _survey_results_public.csv_ file, acting like a schema.

````python
# Load schema from csv file.
sdf = pd.read_csv("./data/Stackoverflow/data/survey_results_schema.csv")
````

#### Modify option `display.max_rows` to display required number of rows

````python
old_max_rows = pd.options.display.max_rows
print("Old value in option `display.max_rows`:", old_max_rows)

(row_count, col_count) = sdf.shape
pd.options.display.max_rows = row_count
print("New value in option `display.max_rows`:", row_count)
````

#### Modify styling to left align column values

Applying style on a dataframe returns a [Styler object](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.io.formats.style.Styler.html#pandas.io.formats.style.Styler), **not a DataFrame**. You cannot apply further style operations on that.

Ref: [Stackoverflow answer.](https://stackoverflow.com/questions/67869034/styler-object-has-no-attribute-style#answer-67869252)

````python
s_sdf = sdf.style.set_properties(**{"text-align": "left"})  # Styled Schema Dataframe
s_sdf
````
